```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection

## Model_selection is used for splitting data into training and test sets
```

## Loading the dataset

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])

# X is the array with the features and Y is the diabetes progression (our targer)
```

```python
X = X[:, 2]  #Here we only take the bmi body mass index
X = X.reshape((-1,1))

## Here we take only the third column to analyze. Then we reshape the data.
```

## Training the model

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
# Test size parameter determines the proportion of the dataset to include
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
